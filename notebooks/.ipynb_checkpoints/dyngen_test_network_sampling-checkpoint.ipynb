```r
create_subnetwork <- function(seed, source_graph, subnet_size) {
    # start of the subnet graph
    subnet_nodes <- c(seed)
    
    # adding nodes to the extracted sub network until i read users specified size
    while (length(subnet_nodes) < subnet_size) {
        subnet_nodes <- add_node_to_subnetwork(subnet_nodes, source_graph)
    }
    
    subnet <- induced_subgraph(source_graph, subnet_nodes)
    return (subnet)
}

add_node_to_subnetwork <- function(subnet_nodes, source_graph) {
    nieghboring_nodes <- get_nieghboring_nodes(subnet_nodes, source_graph)
    nieghbor_df <- calculate_nieghbor_modularity(nieghboring_nodes, subnet_nodes, source_graph)
    node_to_add <- get_node_with_highest_modularity(nieghbor_df, subnet_nodes)
    
    subnet_nodes <- c(subnet_nodes, node_to_add)
    return (subnet_nodes)
}

get_nieghboring_nodes <- function(subnet_nodes, source_graph) {
    nieghbor_nodes <- c()
    
    for (node in subnet_nodes){
        current_nieghbors <- neighbors(source_graph, c(node), mode='out')$name
        nieghbor_nodes <- c(nieghbor_nodes, current_nieghbors)
    }
    
    nieghbor_nodes <- unique(nieghbor_nodes)
    return (nieghbor_nodes)
}

calculate_nieghbor_modularity <- function(nieghbor_nodes, subnet_nodes, source_graph) {
    nieghbor_df <- data.frame(nodes=character(), modularity=double(), stringsAsFactors=FALSE)
    
    for (nieghbor in nieghbor_nodes) {
        temp_nodes <- c(subnet_nodes, nieghbor)
        membership_ids <- extract_subnet_membership(temp_nodes, names(V(source_graph)))

        q <- modularity(source_graph, membership_ids)
        nieghbor_df <- nieghbor_df %>% add_row(nodes=nieghbor, modularity=q)
    }
    
    return (nieghbor_df)
}

get_node_with_highest_modularity <- function(nieghbor_df, subnet_nodes){
    node_to_add <- nieghbor_df %>% filter(!nodes %in% subnet_nodes & modularity == max(modularity)) %>% pull(nodes)
    node_to_add <- ifelse(length(node_to_add) == 1, node_to_add, sample(node_to_add))
    
    return (node_to_add)
}

extract_subnet_membership <- function(temp_nodes, regulators){   
    membership_id <- c()
    
    for (node in regulators) {
        if (node %in% temp_nodes){
            membership_id <- c(membership_id, 2)
        } else {
            membership_id <- c(membership_id, 1)
        }
    }
    
    return (membership_id)
}
```

```r
temp <- fantom5[['regulatorycircuits_01_neurons_fetal_brain']]
```

```r
source_graph <- temp %>% 
                    Matrix::summary() %>% 
                    as.data.frame() %>% 
                    group_by(.data$j) %>% 
                    ungroup() %>% 
                    transmute(
                      i = rownames(reg_net)[.data$i], 
                      j = colnames(reg_net)[.data$j],
                      weight = .data$x) %>%
                    graph_from_data_frame(vertices = colnames(reg_net))
```

```r
subnet_size = 20
min_out_degree <- 5
reg_nodes <- names(V(source_graph))
regs_to_sample <- degree(source_graph, v = reg_nodes, mode = "out") %>%
                    as.data.frame %>%
                    filter(. > min_out_degree) %>%
                    rownames(.)

seed_nodes <- sample(regs_to_sample, size=10)
```

```r
insilico_grns <- list()

for (seed in seed_nodes) {
    subnet <- create_subnetwork(seed, source_graph, subnet_size)
    subnet_name <- paste0('subnetwork_from_', seed)
    
    insilico_grns[[subnet_name]]$subnet <- subnet
    insilico_grns[[subnet_name]]$subnet_name <- subnet_name
}
```
