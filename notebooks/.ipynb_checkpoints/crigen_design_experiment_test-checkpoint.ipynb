```r
sample_percentage = 0.5
n_rows = nrow(sim_edgelist)
sample_size = round(n_rows * sample_percentage)

sampled_edges = sample(1:n_rows, size=sample_size)
sampled_edgelist = sim_edgelist[sampled_edges, ]
```

```r
sampled_net = graph_from_edgelist(sampled_edgelist)
sampled_phi = as.matrix(as_adjacency_matrix(sampled_net))
```
