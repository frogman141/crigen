```r
infer_phi <- as(b$graph, 'matrix')
infer_phi <- (infer_phi > 0.5) * 1
diff_phi <- true_phi - infer_phi

tp <- sum(true_phi & infer_phi)
tn <- sum(!true_phi & !infer_phi) - length(diag(true_phi))
fp <- sum(diff_phi < 0)
fn <- sum(diff_phi > 0)

sensitivity <- tp / (tp + fn)
specificity <- tn / (tn + fp)
balanced_accuracy <- (sensitivity + specificity) / 2

info_dt <- data.table('time' = time_diff,
                    'sensitivity' = sensitivity,
                    'specificity' = specificity,
                    'balanced_accuracy' = balanced_accuracy,
                    additional_input_args)
```
